#1.Write a Python program to create a basic calculator that performs addtions,substraction,multiplication,and division based on user input

```python
no_1=float(input("Enter your first number:"))
no_2=float(input("Enter your second number:"))
operator=input("Enter the operator,(+,-,*,/):")
if operator == '+':
             print('result:',no_1+no_2)
elif operator=='-':
            print('result:',no_1-no_2)
elif operator=='*':
            print('result:',no_1*no_2)
elif operator=='/':
            print('result:',no_1/no_2)
else:
    print("Enter valid operator")
```
